## Clustering Algorithms: K-Means

K-Means is the most basic of unsupervised clustering algorithms, very useful even in business settings, such as to do customer segmentation, basic fraud/anomaly detection, etc..

In this notebook, I will go through the steps required to implement the K-Means algorithm, as well as certain techniques to improve performance.

Although there are already existing libraries (SK-Learn) that have a really efficient implementation, building it from scratch will give much better inuitions on its strengths and shortcomings.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster, datasets
#print(plt.style.available)
plt.style.use("ggplot")
%matplotlib inline
```

```python
n_samples = 1500
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=.5,noise=.05)[0]
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)[0]
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)[0]
```

```python
plt.scatter(x=blobs[:,0],y=blobs[:,1],c="black")
plt.title("How many clusters are there?")
```

# The 5 Steps in K-Means Algorithm

```python
def plot_cluster(data,centroids,cluster):
    ## Plot cluster and centroids
        plt.scatter(x=data[:,0],y=data[:,1],c=cluster)
        plt.scatter(x=centroids[:,0],y=centroids[:,1],s=80, c='r', marker='o',)
```

```python
def initialize_centroids(data,num_clusters):
    centroids = data[np.random.choice(data.shape[0],size=num_clusters,replace=False)]
    return centroids
```

```python
np.random.seed(seed=1)
centroids = initialize_centroids(blobs, 3)
plot_cluster(blobs,centroids,np.zeros(len(centroids)))
plt.title("Step 1: Randomly Initialize Centroids")
```

```python
def assign_cluster(data,centroids,num_clusters):
    ## Initialize array of zeros to hold distances of each point from centroids
    dist = np.zeros(shape=(data.shape[0],num_clusters), dtype=np.float32)
    ## Calculate distance from centroids - non-vectorized
    for i_d , d in enumerate(data):
        for i_c, c in enumerate(centroids):
            sq_distance = 0
            for i in range(len(d)):
                sq_distance += (d[i]-c[i])**2
            dist[i_d,i_c] = sq_distance ** (1/2)
    ## Assign each point to a cluster centroid nearest to it
    cluster = np.argmin(dist,axis=1)
    return cluster
```

```python
cluster = assign_cluster(blobs,centroids,3)
plot_cluster(blobs,centroids,cluster)
plt.title("Step 2: Assign to Cluster based on Initial Centroids")
```

While the above function to assign clusters works, it uses 3 for-loops to achieve its objective. 

This means that the runtime of the code will increase exponentially as any of the following variables increase:
 - Number of points in dataset;
 - Number of features of each datapoint; 
 - Number of centroids
 
For large, high dimension datasets, the code will take forever to run. Switch to a vectorized version that runs much faster.

```python
def assign_cluster_v(data,centroids,num_clusters):
    ## Calcuate distance between each data point and centroids
    sq_diff = np.square(data[:,np.newaxis]-centroids)
    dist = np.sqrt(np.sum(sq_diff,axis=2))
    ## Assign each point to a cluster centroid nearest to it
    cluster = np.argmin(dist,axis=1)
    return cluster
```

```python
def update_centroids(data,centroids,cluster):
    ## Update new centroids
    new_centroids = np.zeros(centroids.shape)
    for i , _ in enumerate(new_centroids):
        new_centroids[i] = np.mean(data[cluster == i],axis=0)
    return new_centroids
```

```python
new_centroids = update_centroids(blobs,centroids,cluster)
plot_cluster(blobs,new_centroids,cluster)
plt.title("Step 3: Update New Centroids with Cluster Mean")
```

```python
cluster = assign_cluster_v(blobs,new_centroids,3)
plot_cluster(blobs,new_centroids,cluster)
plt.title("Step 4: Assign to Cluster based on New Centroids")
```

```python
switch = True
while switch:
    old_centroids = new_centroids
    new_centroids = update_centroids(blobs,old_centroids,cluster)
    cluster = assign_cluster_v(blobs,new_centroids,3)
    ## If centroids stop changing position, training completed.
    if np.array_equal(old_centroids,new_centroids):
        switch = False
plot_cluster(blobs,new_centroids,cluster)
plt.title("Step 5: Repeat Step 3&4 until centroids stop moving")
```

```python
## Package into a single k_means function
def k_means(data,num_clusters,plot=True):
    new_centroids = initialize_centroids(data,num_clusters)
    cluster = assign_cluster_v(data,new_centroids,num_clusters)
    switch = True
    while switch:
        old_centroids = np.copy(new_centroids)
        new_centroids = update_centroids(data,old_centroids,cluster)
        cluster = assign_cluster_v(data,new_centroids,num_clusters)
        ## If centroids stop changing position, training completed.
        if np.array_equal(old_centroids,new_centroids):
            switch = False
    if plot:
        plot_cluster(data,new_centroids,cluster)
    return (cluster, new_centroids)
```

# Test K_means against other shapes of data

```python
np.random.seed(1)
k_means(noisy_moons,2)
plt.title("K-Means, 2 Clusters.")
```

```python
np.random.seed(1)
k_means(noisy_circles,2)
plt.title("K-Means, 2 Clusters.")
```

# The importance of initialization in K-Means

```python
np.random.seed(4)
k_means(blobs,3)
plt.title("K-Means, 3 Clusters. What happened??")
```

```python
np.random.seed(seed=4)
centroids = initialize_centroids(blobs, 3)
plot_cluster(blobs,centroids,np.zeros(len(centroids)))
plt.title("Answer: Unlucky/Bad Initialization!")
```

# Comparing Vectorized vs Non-Vectorized Implementation of K-Means

```python
## The K_Means Non_vectorized Version
def k_means_nonv(data,num_clusters,plot=True):
    new_centroids = initialize_centroids(data,num_clusters)
    cluster = assign_cluster(data,new_centroids,num_clusters)
    switch = True
    while switch:
        old_centroids = np.copy(new_centroids)
        new_centroids = update_centroids(data,old_centroids,cluster)
        cluster = assign_cluster(data,new_centroids,num_clusters)
        ## If centroids stop changing position, training completed.
        if np.array_equal(old_centroids,new_centroids):
            switch = False
    if plot:
        plot_cluster(data,new_centroids,cluster)
    return (cluster, new_centroids)
```

```python
## Sample data with 100,000 points
large_blob = datasets.make_blobs(n_samples=100000, random_state=8)[0]
```

```python
from time import time
```

## 1. Our K-Means Implemention (Non-Vectorized)

```python
np.random.seed(1)
tic = time()
cluster, centroids = k_means_nonv(large_blob,3)
toc = time()
elapsed = (toc - tic)
print("Time Elapsed: {:2f} seconds".format(elapsed))
plot_cluster(large_blob,centroids,cluster)
```

## 2. Our K-Means Implemention (Vectorized)

```python
np.random.seed(1)
tic = time()
cluster, centroids = k_means(large_blob,3)
toc = time()
elapsed = (toc - tic)
print("Time Elapsed: {:.2f} seconds".format(elapsed))
plot_cluster(large_blob,centroids,cluster)
```

## 3. SK-Learn K-Means Implemention

```python
## Lets compare our implementation to sklearn's
from sklearn.cluster import KMeans
np.random.seed(1)
tic = time()
model = KMeans(n_clusters=3,n_init=1).fit(large_blob)
y_pred = model.predict(large_blob)
toc = time()
elapsed = (toc - tic)
print("Time Elapsed: {:2f} seconds".format(elapsed))
plot_cluster(large_blob,model.cluster_centers_,y_pred)
```
